```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Cars Company.csv")
df
```

#### 1. Check and Drop Duplicate Records

```python
df = df.drop_duplicates()
df
```

#### 2. Check and Drop missing or null values

```python
df = df.dropna()
df
```

#### 3. What are the top 10 average MSRP (Price) per car brand (rounded figure)? with Horizontal Bar Plot

```python
top_10_avg_msrp = df.groupby('Make')['MSRP'].mean().round().sort_values(ascending=False).head(10)
top_10_avg_msrp.plot(kind='barh', figsize=(10,6), color='skyblue')
plt.xlabel('Average MSRP ($)')
plt.title('Top 10 Average MSRP per Car Brand')
plt.gca().invert_yaxis()
plt.show()
```

#### 4. How has the MSRP (price) changed over the years (rounded figure)? with Line Plot

```python
msrp_by_year = df.groupby('Year')['MSRP'].mean().round()
msrp_by_year.plot(kind='line', marker='o', figsize=(10,6))
plt.ylabel('Average MSRP ($)')
plt.title('Average MSRP Over the Years')
plt.grid(True)
plt.show()
```

#### 5. How Engine HP (power) and MPG (efficiency) have changed over the years? show with two line plots

Hint: Average of (city mpg + highway MPG)

```python
df['Average_MPG'] = (df['city mpg'] + df['highway MPG']) / 2
engine_hp_trend = df.groupby('Year')['Engine HP'].mean()
mpg_trend = df.groupby('Year')['Average_MPG'].mean()

plt.figure(figsize=(10,6))
plt.plot(engine_hp_trend, marker='o', label='Average Engine HP')
plt.plot(mpg_trend, marker='s', label='Average MPG')
plt.title('Engine HP and MPG Over the Years')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

#### 6. What percentage of cars are Compact, Midsize, or Large? Show with Pie chart

```python
size_counts = df['Vehicle Size'].value_counts()
size_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6,6))
plt.title('Vehicle Size Distribution')
plt.ylabel('')
plt.show()
```

#### 7. Which vehicle styles are sold the most? show with vertical bar plot

```python
style_counts = df['Vehicle Style'].value_counts()
style_counts.plot(kind='bar', figsize=(10,6), color='coral')
plt.ylabel('Number of Vehicles')
plt.title('Most Sold Vehicle Styles')
plt.xticks(rotation=45)
plt.show()
```

#### 8. Which car brands have the most vehicles priced at $40,000 or more and with a popularity score of at least 1000?

```python
filtered = df[(df['MSRP'] >= 40000) & (df['Popularity'] >= 1000)]
brand_counts = filtered['Make'].value_counts()
print("Brands with most high-value & popular vehicles:")
print(brand_counts)
```
